```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
dataset = keras.datasets.imdb
(train_data, train_labels), (test_data, test_labels) = dataset.load_data(num_words = 10000)
```

```python
def vectorize(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequences in enumerate(sequences):
        results[i, sequences] = 1.
    return results

x_train = vectorize(train_data)
x_test = vectorize(test_data)
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')



x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

```python
# Standardize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(partial_x_train)
x_val_scaled = scaler.transform(x_val)
x_test_scaled = scaler.transform(x_test)

# Hyperparameter tuning for kNN
k_values = [1, 3, 5, 7, 9]
val_accuracies_knn = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train_scaled, partial_y_train)
    y_val_pred = knn.predict(x_val_scaled)
    val_accuracy = accuracy_score(y_val_pred, y_val)
    val_accuracies_knn.append(val_accuracy)
    print(f'k = {k}: Validation Accuracy = {val_accuracy}')

# Plot validation accuracy for different k values
plt.figure(figsize=(10, 6))
plt.plot(k_values, val_accuracies_knn, marker='o')
plt.title('Validation Accuracy for kNN Classifier with Different k Values')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Validation Accuracy')
plt.grid(True)
plt.show()

# Test accuracy with the best k
best_k = k_values[val_accuracies_knn.index(max(val_accuracies_knn))]
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(x_train_scaled, partial_y_train)
knn_test_accuracy = knn_best.score(x_test_scaled, y_test)
print(f'Best k = {best_k}: Test Accuracy = {knn_test_accuracy}')
```

```python
# Hyperparameter tuning for Logistic Regression
C_values = [0.01, 0.1, 1, 10, 100]
val_accuracies_log_reg = []

for C in C_values:
    log_reg = LogisticRegression(C=C, max_iter=1000)
    log_reg.fit(x_train_scaled, partial_y_train)
    y_val_pred = log_reg.predict(x_val_scaled)
    val_accuracy = accuracy_score(y_val_pred, y_val)
    val_accuracies_log_reg.append(val_accuracy)
    print(f'C = {C}: Validation Accuracy = {val_accuracy}')

# Plot validation accuracy for different C values
plt.figure(figsize=(10, 6))
plt.plot(C_values, val_accuracies_log_reg, marker='o')
plt.title('Validation Accuracy for Logistic Regression with Different C Values')
plt.xlabel('Regularization Strength (C)')
plt.ylabel('Validation Accuracy')
plt.xscale('log')
plt.grid(True)
plt.show()

# Test accuracy with the best C
best_C = C_values[val_accuracies_log_reg.index(max(val_accuracies_log_reg))]
log_reg_best = LogisticRegression(C=best_C, max_iter=1000)
log_reg_best.fit(x_train_scaled, partial_y_train)
log_reg_test_accuracy = log_reg_best.score(x_test_scaled, y_test)
print(f'Best C = {best_C}: Test Accuracy = {log_reg_test_accuracy}')
```
